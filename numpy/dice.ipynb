```python
##dice rolling##
print("Welcome to the dice rolling simulator!")
import random
def roll_dice():
    while True:
        try:
            sides = int(input("Enter the number of sides on the dice (6, 10, 20, etc.): "))
            if sides < 1:
                print("Please enter a positive integer for the number of sides.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

    roll = random.randint(1, sides)
    print(f"You rolled a {roll} on a {sides}-sided dice.")
    
```

```python
##guess number
def guess_number():
    number_to_guess = random.randint(1, 100)
    attempts = 0
    print("Guess a number between 1 and 100.")

    while True:
        try:
            guess = int(input("Enter your guess: "))
            attempts += 1
            if guess < 1 or guess > 100:
                print("Please guess a number between 1 and 100.")
                continue
            if guess < number_to_guess:
                print("Too low! Try again.")
            elif guess > number_to_guess:
                print("Too high! Try again.")
            else:
                print(f"Congratulations! You guessed the number {number_to_guess} in {attempts} attempts.")
                break
        except ValueError:
            print("Invalid input. Please enter a valid integer.")
            
```

```python
print("Welcome to the Basic To-Do List!")

# Step 1: Get initial items
list1 = input("Enter your to-do list items separated by commas: ")
todo_list = list1.split(",")

# Clean up spaces
todo_list = [item.strip() for item in todo_list]

# Step 2: Show menu
def show_menu():
    print("\nChoose an option:")
    print("1. View To-Do List")
    print("2. Add Item")
    print("3. Remove Item")
    print("4. Exit")

# Step 3: Main loop
while True:
    show_menu()
    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        print("\nYour To-Do List:")
        for i, item in enumerate(todo_list, start=1):
            print(f"{i}. {item}")

    elif choice == "2":
        new_item = input("Enter the item to add: ")
        todo_list.append(new_item.strip())
        print(f'"{new_item}" added to the list.')

    elif choice == "3":
        remove_item = input("Enter the item to remove: ")
        if remove_item.strip() in todo_list:
            todo_list.remove(remove_item.strip())
            print(f'"{remove_item}" removed from the list.')
        else:
            print("Item not found in the list.")

    elif choice == "4":
        print("Goodbye! ✅")
        break

    else:
        print("Invalid choice. Please enter 1–4.")
def show_menu():
    print("\nChoose an option:")
    print("1. Roll Dice")
    print("2. Guess Number")
    print("3. To-Do List")
    print("4. Exit")

```
